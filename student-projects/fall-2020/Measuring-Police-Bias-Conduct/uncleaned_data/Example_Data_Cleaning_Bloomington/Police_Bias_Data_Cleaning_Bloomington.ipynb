# Bloomington Data Cleaning

## Use of Force

Read in the datasets.

### 2019

```python
uofB1_19 = pd.read_excel (r'Use of Force/Bloomington/2019/2019-first-quarter-use-of-force.xlsx')
uofB2_19 = pd.read_excel (r'Use of Force/Bloomington/2019/2019-second-quarter-use-of-force.xlsx')
uofB3_19 = pd.read_excel (r'Use of Force/Bloomington/2019/2019-third-quarter-use-of-force.xlsx')
uofB4_19 = pd.read_excel (r'Use of Force/Bloomington/2019/2019-fourth-quarter-use-of-force.xlsx')

uof_19 = pd.concat([uofB1_19, uofB2_19, uofB3_19, uofB4_19])
uof_19 = uof_19.drop(columns = ['Draw Less Lethal'])
```

### 2018

```python
uofB1_18 = pd.read_excel (r'Use of Force/Bloomington/2018/2018-first-quarter-use-of-force.xlsx')
uofB2_18 = pd.read_excel (r'Use of Force/Bloomington/2018/2018-second-quarter-use-of-force.xlsx')
uofB3_18 = pd.read_excel (r'Use of Force/Bloomington/2018/2018-third-quarter-use-of-force.xlsx')
uofB4_18 = pd.read_excel (r'Use of Force/Bloomington/2018/2018-fourth-quarter-use-of-force.xlsx')

uof_18 = pd.concat([uofB1_18, uofB2_18, uofB3_18, uofB4_18])
```

### 2017

```python
uofB1_17 = pd.read_excel (r'Use of Force/Bloomington/2017/2017-first-quarter-use-of-force.xlsx')
uofB2_17 = pd.read_excel (r'Use of Force/Bloomington/2017/2017-second-quarter-use-of-force.xlsx')
uofB3_17 = pd.read_excel (r'Use of Force/Bloomington/2017/2017-third-quarter-use-of-force.xlsx')
uofB4_17 = pd.read_excel (r'Use of Force/Bloomington/2017/2017-fourth-quarter-use-of-force.xlsx')

uof_17 = pd.concat([uofB1_17, uofB2_17, uofB3_17, uofB4_17])
```

### 2016

```python
uofB1_16 = pd.read_excel (r'Use of Force/Bloomington/2016/2016-first-quarter-use-of-force.xlsx')
uofB2_16 = pd.read_excel (r'Use of Force/Bloomington/2016/2016-second-quarter-use-of-force.xlsx')
uofB3_16 = pd.read_excel (r'Use of Force/Bloomington/2016/2016-third-quarter-use-of-force.xlsx')
uofB4_16 = pd.read_excel (r'Use of Force/Bloomington/2016/2016-fourth-quarter-use-of-force.xlsx')

uof_16 = pd.concat([uofB1_16, uofB2_16, uofB3_16, uofB4_16])
uof_16 = uof_16.reset_index()
uof_16 = uof_16.drop([0, 293])
uof_16['Date'] = pd.to_datetime(uof_16['Date']).dt.date
```

### Combine all years into one DataFrame

```python
uof_all_years = pd.concat([uof_19, uof_18, uof_17, uof_16]).reset_index().drop(columns = ['index'])
```

### Fill null values with 0 for quantitative categories and Uknown for qualitative categories
* The null values here represent that that suspect was not of that race, meaning that there is a 1 in another category of his/her respective race. So to be able to tell that they are not of that race, instead we fill it with the value of 0.

```python
uof_all_years = uof_all_years.fillna(value = {'Suspect Gender Female' : 0, 'Suspect Gender Male': 0, 
                                              'Suspect Race  Caucasian, Hispanic': 0, 
                                              'Suspect Race Caucasian, Non-Hispanic': 0, 
                                              'Suspect Race African American, Non-Hispanic': 0, 
                                              'Suspect Race African American, Hispanic': 0, 
                                              'Suspect Race Indian/Alaskan Native, Non-Hispanic': 0, 
                                              'Suspect Race Indian/Alaskan Native, Hispanic': 0,
                                              'Suspect Race Asian/Pacific Islander, Non-Hispanic': 0, 
                                              'Suspect Race Asian/Pacific Islander, Hispanic': 0, 
                                              'District': 'Unknown', 'Suspect Condition': 'Unknown', 
                                              'Suspect Role': 'Unknown' })

uof_all_years['Year'] = pd.DatetimeIndex(uof_all_years['Date']).year
uof_all_years = uof_all_years.drop(['level_0'], axis = 1)
```

```python
white_uof_19 = uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race Caucasian, Non-Hispanic'].sum()
black_uof_19 = uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race African American, Non-Hispanic'].sum()
latinx_uof_19 = uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race  Caucasian, Hispanic'].sum()
asian_uof_19 = uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race Asian/Pacific Islander, Non-Hispanic'].sum()
other_uof_19 = uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race Indian/Alaskan Native, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race Indian/Alaskan Native, Non-Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race Asian/Pacific Islander, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2019]['Suspect Race African American, Hispanic'].sum()

white_uof_18 = uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race Caucasian, Non-Hispanic'].sum()
black_uof_18 = uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race African American, Non-Hispanic'].sum()
latinx_uof_18 = uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race  Caucasian, Hispanic'].sum()
asian_uof_18 = uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race Asian/Pacific Islander, Non-Hispanic'].sum()
other_uof_18 = uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race Indian/Alaskan Native, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race Indian/Alaskan Native, Non-Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race Asian/Pacific Islander, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2018]['Suspect Race African American, Hispanic'].sum()

white_uof_17 = uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race Caucasian, Non-Hispanic'].sum()
black_uof_17 = uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race African American, Non-Hispanic'].sum()
latinx_uof_17 = uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race  Caucasian, Hispanic'].sum()
asian_uof_17 = uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race Asian/Pacific Islander, Non-Hispanic'].sum()
other_uof_17 = uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race Indian/Alaskan Native, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race Indian/Alaskan Native, Non-Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race Asian/Pacific Islander, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2017]['Suspect Race African American, Hispanic'].sum()


white_uof_16 = uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race Caucasian, Non-Hispanic'].sum()
black_uof_16 = uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race African American, Non-Hispanic'].sum()
latinx_uof_16 = uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race  Caucasian, Hispanic'].sum()
asian_uof_16 = uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race Asian/Pacific Islander, Non-Hispanic'].sum()
other_uof_16 = uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race Indian/Alaskan Native, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race Indian/Alaskan Native, Non-Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race Asian/Pacific Islander, Hispanic'].sum() 
+ uof_all_years[uof_all_years['Year'] == 2016]['Suspect Race African American, Hispanic'].sum()

```

## Stop Data / Citations

Read in the datasets.

### 2016 - 2019

```python
cit_19 = pd.read_excel (r'Citations/Bloomington/2019.xlsx')
cit_18 = pd.read_excel (r'Citations/Bloomington/2018.xlsx')
cit_17 = pd.read_excel (r'Citations/Bloomington/2017.xlsx')
cit_16 = pd.read_excel (r'Citations/Bloomington/2016.xlsx')
```

### Fill null values with Unknown

```python
cit_19 = cit_19.fillna('Unknown')
cit_18 = cit_18.fillna('Unknown')
cit_17 = cit_17.fillna('Unknown')
cit_16 = cit_16.fillna('Unknown')
```

### Merge all years together

```python
cit_19['Year'] = 2019
cit_18['Year'] = 2018
cit_17['Year'] = 2017
cit_16['Year'] = 2016
cit_bloom_merged = pd.concat([cit_19, cit_18, cit_17])
cit_bloom_merged
cit_bloom_merged.to_csv('all_cities.csv')
cit_16.to_csv('bloom_16.csv')
cit_bloom_merged = cit_bloom_merged.reset_index()
cit_bloom_merged
```

### Get the total number of persons by race per year

```python
black_drive_stops_19 = sum([(cit_bloom_merged['Cited Person Race'] == 'B') & (cit_bloom_merged['Year'] == 2019)]).sum()   
white_drive_stops_19 = sum([(cit_bloom_merged['Cited Person Race'] == 'W') & (cit_bloom_merged['Year'] == 2019)]).sum()   
latinx_drive_stops_19 = sum([(cit_bloom_merged['Cited Person Race'] == 'L') & (cit_bloom_merged['Year'] == 2019)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'K') & (cit_bloom_merged['Year'] == 2019)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'N') & (cit_bloom_merged['Year'] == 2019)]).sum()  
asian_drive_stops_19 = sum([(cit_bloom_merged['Cited Person Race'] == 'A') & (cit_bloom_merged['Year'] == 2019)]).sum()     
other_drive_stops_19 = sum([(cit_bloom_merged['Cited Person Race'] == 'U') & (cit_bloom_merged['Year'] == 2019)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'I') & (cit_bloom_merged['Year'] == 2019)]).sum()

black_drive_stops_18 = sum([(cit_bloom_merged['Cited Person Race'] == 'B') & (cit_bloom_merged['Year'] == 2018)]).sum()   
white_drive_stops_18 = sum([(cit_bloom_merged['Cited Person Race'] == 'W') & (cit_bloom_merged['Year'] == 2018)]).sum()
latinx_drive_stops_18 = sum([(cit_bloom_merged['Cited Person Race'] == 'L') & (cit_bloom_merged['Year'] == 2018)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'K') & (cit_bloom_merged['Year'] == 2018)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'N') & (cit_bloom_merged['Year'] == 2018)]).sum()
asian_drive_stops_18 = sum([(cit_bloom_merged['Cited Person Race'] == 'A') & (cit_bloom_merged['Year'] == 2018)]).sum()     
other_drive_stops_18 = sum([(cit_bloom_merged['Cited Person Race'] == 'U') & (cit_bloom_merged['Year'] == 2018)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'I') & (cit_bloom_merged['Year'] == 2018)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'K') & (cit_bloom_merged['Year'] == 2018)]).sum()

black_drive_stops_17 = sum([(cit_bloom_merged['Cited Person Race'] == 'B') & (cit_bloom_merged['Year'] == 2017)]).sum() 
white_drive_stops_17 = sum([(cit_bloom_merged['Cited Person Race'] == 'W') & (cit_bloom_merged['Year'] == 2017)]).sum() 
latinx_drive_stops_17 = sum([(cit_bloom_merged['Cited Person Race'] == 'L') & (cit_bloom_merged['Year'] == 2017)]).sum() + sum([(cit_bloom_merged['Cited Person Race'] == 'K') & (cit_bloom_merged['Year'] == 2017)]).sum()  
asian_drive_stops_17 = sum([(cit_bloom_merged['Cited Person Race'] == 'A') & (cit_bloom_merged['Year'] == 2017)]).sum()    
other_drive_stops_17 = sum([(cit_bloom_merged['Cited Person Race'] == 'P') & (cit_bloom_merged['Year'] == 2017)]).sum()    
+ sum([(cit_bloom_merged['Cited Person Race'] == 'P') & (cit_bloom_merged['Year'] == 2017)]).sum()    
+ sum([(cit_bloom_merged['Cited Person Race'] == 'P') & (cit_bloom_merged['Year'] == 2017)]).sum()    

black_drive_stops_16 = sum(cit_16['Cited Person Race'] == 'B')
white_drive_stops_16 = sum(cit_16['Cited Person Race'] == 'W')
latinx_drive_stops_16 = sum(cit_16['Cited Person Race'] == 'L') + sum(cit_16['Cited Person Race'] == 'K')
asian_drive_stops_16 = sum(cit_16['Cited Person Race'] == 'A')
other_drive_stops_16 = sum(cit_16['Cited Person Race'] == 'U') +sum(cit_16['Cited Person Race'] == 'P')
```

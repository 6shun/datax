```python
import csv
import pandas as pd 
import numpy as np
infile = 'lol_chatlog4.csv'
bloblist = list()

with open(infile, 'r',encoding='UTF-8') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        sentence = row[3]
        blob = TextBlob(sentence)
        print (sentence)
        print (blob.sentiment.polarity, blob.sentiment.subjectivity)
        bloblist.append((sentence,blob.sentiment.polarity, blob.sentiment.subjectivity))
```

```python
len(bloblist)
```

```python
df = pd.DataFrame(bloblist, columns = ['sentence','sentiment','polarity'])
df[90:110]
```
